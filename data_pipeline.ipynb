```python
import pandas as pd
```

```python
df = pd.read_csv("test_data.csv")
```

```python
df1 = df.fillna("")
```

```python
df1
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("test_data.db")
```

```python
df1.to_sql("test_data", conn, if_exists="replace")
```

```python
conn.execute(
    """
    create table emp as 
    select * from test_data
    """)
```

```python
df2 = pd.read_sql_query("SELECT * from emp", conn)
```

```python
df2
```

```python
df_KNK = df2.groupby("Kwt_Nk").size()
```
