```python
import pandas as pd
```

```python
df = pd.read_csv("test_data.csv", encoding ='latin1')
```

```python
df
```

```python
df1 = df.fillna("")
```

```python
df1
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("test_data.db")
```

```python
df1.to_sql("test_data", conn, if_exists="replace")
```

```python
conn.execute(
    """
    create table emp_new as 
    select * from test_data
    """)
```

```python
df2 = pd.read_sql_query("SELECT * from emp_new", conn)
```

```python
df2
```

```python
df_KNK = df2.groupby("Kwt_Nk").size()
```

```python
df_KNK
```

```python
df_Gender = df2.groupby("Gender").size()
```

```python
df_Gender
```

```python
df_mainos = df2.groupby(['main_os', 'Kwt_Nk'], as_index=False).count()
```
